<h1>PandaSQL</h1>

```python
def pandasql(d1, d2):
    join_query = '''
        SELECT 
            *
        FROM
            d1 
            JOIN 
            d2
                on d1.use_id = d2.use_id
    '''
    return ps.sqldf(join_query, locals())
```

```python
pandasql(d1, d2)
```

```python
def pandasql(d2):
    aggr_query = '''
        SELECT 
            avg(outgoing_mins_per_month) as outgoing_mins_per_month,
            platform
        FROM d2 
        GROUP BY platform
        '''
    return ps.sqldf(aggr_query, locals()).set_index('platform')
```
